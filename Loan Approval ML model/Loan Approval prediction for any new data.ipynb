Import necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score, precision_score, recall_score
import matplotlib.pyplot as plt
from joblib import load
```


Load the trained model

```python
model_t = load('saved model.joblib') #load the trained machine
```


Read the new data from a file that needed to be tested

```python
new_file = input("Hi there! Please provide the name of the data file you want the model to predict:")
new_dataset = pd.read_csv(f"{new_file}.csv")
pd.DataFrame(new_dataset)
```


Read the original dataset to help with the new data prediction


```python
original_file = input("Now, please provide the name of the original data file the model used for previous predictions:")
df = pd.read_csv(f"{original_file}.csv")
```


Drop the unnecessary colums from both data sets where it can cuase for biases


```python
x = df.drop(columns=["Loan Approval", "Gender","Age"],axis = 1)
new_dataset = new_dataset.drop(columns=["Gender","Age"],axis = 1)

#convert columns into multiple binary columns
new_dataset = pd.get_dummies(new_dataset, drop_first= True)
x = pd.get_dummies(x)

pd.DataFrame(new_dataset)
```

Calculates how many missing colums and iterate through the colums and define the empty ones  with zeros

```python

missing_cols = set(x.columns) - set(new_dataset.columns)

for col in missing_cols:
    new_dataset[col] = 0

#make sure that the new data features and the original data feautres are in the same order     
new_dataset = new_dataset[x.columns]
```

Using the model already trained, predict if the customer eligible for getting a loan

```python
pred = model_t.predict(new_dataset)
```

Save the file in the device

```python
new_dataset['Loan Approval'] = pred

#Save the data as a csv file and save it
path = input("Where would you like to save the file at?\n")
file_name = input("What would you like the file be named as?\n")

save = rf"{path}\{file_name}.csv"

new_dataset.to_csv(save, index = False)
```
